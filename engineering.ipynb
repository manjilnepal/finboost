```python
# Borrow-to-Deposit ratio
df_train['borrow_to_deposit'] = np.where(df_train['userDepositSum'] > 0,
                                   df_train['userBorrowSum'] / df_train['userDepositSum'], 0)

# Liquidation-to-Borrow ratio
df_train['liquidation_to_borrow'] = np.where(df_train['userBorrowSum'] > 0,
                                       df_train['userLiquidationSum'] / df_train['userBorrowSum'], 0)

# Repay-to-Borrow ratio
df_train['repay_to_borrow'] = np.where(df_train['userBorrowSum'] > 0,
                                 df_train['userRepaySum'] / df_train['userBorrowSum'], 0)

# clean up infinities and NaNs
df_train.replace([np.inf, -np.inf], 0, inplace=True)
df_train.fillna(0, inplace=True)
```
