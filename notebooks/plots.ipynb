```python
import matplotlib.pyplot as plt
import numpy as np

# LaTeX-style formatting (but without needing LaTeX installed)
plt.rcParams.update({
    'font.size': 12,
    'font.family': 'serif',
    'text.usetex': False,
    'axes.linewidth': 1.2,
    'xtick.major.width': 1.2,
    'ytick.major.width': 1.2,
    'grid.linewidth': 0.8
})

# Model names (x-axis)
models = ['CoxPH', 'AFT', 'XGBoost', 'LightGBM', 'DeepSurv']
cindex_values = [0.68, 0.69, 0.84, 0.78, 0.71]  # Example C-index values

# Create figure
fig, ax = plt.subplots(figsize=(6, 4), dpi=600)

# Colors for each model (consistent with previous palette)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#9467bd', '#d62728']

# Bar chart
x = np.arange(len(models))
bars = ax.bar(x, cindex_values, color=colors, edgecolor='black', linewidth=0.9, alpha=0.9)

# Add value labels on top of bars
for i, val in enumerate(cindex_values):
    ax.text(i, val + 0.005, f'{val:.2f}', ha='center', va='bottom',
            fontsize=12, fontweight='bold', color='black')

# Axis labels and title
ax.set_xlabel('Models', fontsize=14, fontweight='bold')
ax.set_ylabel('C-index', fontsize=14, fontweight='bold')

# X ticks
ax.set_xticks(x)
ax.set_xticklabels(models, fontsize=12, fontweight='bold', rotation=70, ha='right')

# Y-axis limits for spacing
ax.set_ylim(0.65, 0.85)

# Gridlines for readability
ax.grid(axis='y', linestyle='--', linewidth=0.8, alpha=0.4, color='gray')
ax.set_axisbelow(True)

# Remove top/right spines
for spine in ['top', 'right']:
    ax.spines[spine].set_visible(False)

# Adjust bottom/left spine aesthetics
for spine in ['left', 'bottom']:
    ax.spines[spine].set_linewidth(1.2)
    ax.spines[spine].set_color('black')

# Tight layout and save
plt.tight_layout()
plt.savefig('cindex_bar.png', dpi=600, bbox_inches='tight', facecolor='white')
plt.show()
```
